```python
import numpy as np
# -----------------------------
# 1Ô∏è‚É£ Create arrays (raw data)
# -----------------------------
farmA = np.array([
    [10, 12, 11],
    [9, 11, 10],
    [12, 13, 14],
    [8, 7, 9]
])

farmB = np.array([
    [20, 19, 21],
    [18, 20, 19],
    [22, 23, 21],
    [17, 18, 16]
])

farmC = np.array([
    [15, 16, 14],
    [14, 13, 15],
    [16, 17, 18],
    [12, 11, 13]
])

# Stack into one 3D array (3 farms √ó 4 panels √ó 3 hours)
production_all = np.stack([farmA, farmB, farmC])

# -----------------------------
# 2Ô∏è‚É£ Array attributes
# -----------------------------
print("Shape:", production_all.shape)
print("Size:", production_all.size)
print("Dtype:", production_all.dtype)
print("Ndim:", production_all.ndim)

# -----------------------------
# 3Ô∏è‚É£ Per-farm totals (sum, mean, std, var)
# -----------------------------
total_per_farm = np.sum(production_all, axis=(1, 2))
mean_per_farm = np.mean(production_all, axis=(1, 2))
std_per_farm = np.std(production_all, axis=(1, 2))
var_per_farm = np.var(production_all, axis=(1, 2))

print("\nTotal per farm:", total_per_farm)
print("Mean per farm:", mean_per_farm)
print("Std per farm:", std_per_farm)
print("Var per farm:", var_per_farm)

# -----------------------------
# 4Ô∏è‚É£ Per-hour weighted estimate
# -----------------------------
hour_weights = np.array([0.5, 1.0, 0.75])
weighted_per_panel = np.dot(production_all, hour_weights)
print("\nWeighted per panel (shape):", weighted_per_panel.shape)
print(weighted_per_panel)

# ----------------------

# -----------------------------
# 6Ô∏è‚É£ Normalization (z-score)
# -----------------------------
normed = (production_all - mean_per_farm[:, None, None]) / std_per_farm[:, None, None]

# -----------------------------
# 7Ô∏è‚É£ Exponential & log family
# -----------------------------
exp_ = np.exp(normed)
expm1_ = np.expm1(normed)
log_ = np.log(np.clip(normed, 1e-6, None))
log10_ = np.log10(np.clip(normed, 1e-6, None))
log2_ = np.log2(np.clip(normed, 1e-6, None))

# -----------------------------
# 8Ô∏è‚É£ Trigonometric
# -----------------------------
sin_ = np.sin(normed)
cos_ = np.cos(normed)
tan_ = np.tan(normed)
arcsin_ = np.arcsin(np.clip(sin_, -1, 1))
arccos_ = np.arccos(np.clip(cos_, -1, 1))
arctan_ = np.arctan(tan_)
deg_ = np.degrees(arctan_)
rad_ = np.radians(deg_)

# -----------------------------
# 9Ô∏è‚É£ Rounding
# -----------------------------
round_ = np.round(normed)
floor_ = np.floor(normed)
ceil_ = np.ceil(normed)
trunc_ = np.trunc(normed)

# -----------------------------
# üîü Comparison & logical
# -----------------------------
high_output = production_all > 20
even_hours_mask = np.array([False, True, False])[None, None, :]
logic_and = np.logical_and(high_output, even_hours_mask)
logic_or  = np.logical_or(high_output, even_hours_mask)
logic_not = np.logical_not(high_output)
logic_xor = np.logical_xor(high_output, even_hours_mask)

ge_ = np.greater_equal(production_all, 20)
lt_ = np.less(production_all, 15)

# -----------------------------
# 11Ô∏è‚É£ Bitwise
# -----------------------------
alerts = np.array([
    [[1,0,1],[0,0,1],[1,1,0],[0,1,0]],  # Farm A
    [[1,1,0],[0,1,1],[1,0,0],[1,1,1]],  # Farm B
    [[0,1,1],[1,1,0],[0,0,1],[1,0,1]]   # Farm C
], dtype=int)

bit_and = np.bitwise_and(alerts, (production_all.astype(int) % 2))
bit_or  = np.bitwise_or(alerts, (production_all.astype(int) % 2))
bit_not = np.invert(alerts)

# -----------------------------
# 12Ô∏è‚É£ Arithmetic ufuncs
# -----------------------------
add_ = np.add(normed, 2)
sub_ = np.subtract(normed, 1)
mul_ = np.multiply(normed, 3)
div_ = np.divide(normed, 2)
pow_ = np.power(normed, 2)
mod_ = np.mod(np.abs(normed*10).astype(int), 4)

# -----------------------------
# 13Ô∏è‚É£ Aggregations on ufunc outputs
# -----------------------------
sum_exp = np.sum(exp_)
mean_sin = np.mean(sin_)

# -----------------------------
# 14Ô∏è‚É£ Broadcast-based summary table
# -----------------------------
panel_totals_per_farm = np.sum(production_all, axis=2)
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)
min_per_farm = np.min(panel_totals_per_farm, axis=1)
max_per_farm = np.max(panel_totals_per_farm, axis=1)

# -----------------------------
# 15Ô∏è‚É£ Final KPIs
# -----------------------------
overall_total = np.sum(production_all)
peak_value = np.max(production_all)
peak_index = np.unravel_index(np.argmax(production_all), production_all.shape)
average_per_panel = np.mean(production_all, axis=(0, 2))

# -----------------------------
# ‚úÖ Print summary
# -----------------------------
print("\n=== Final KPIs ===")
print("Overall total:", overall_total)
print("Peak value:", peak_value, "at farm/panel/hour indices", peak_index)
print("Average per panel:", average_per_panel)
print("Needs maintenance:", needs_maintenance)
print("Sum of exp(normed):", sum_exp)
print("Mean of sin(normed):", mean_sin)
```
